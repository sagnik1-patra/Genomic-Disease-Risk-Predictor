```python
import os
import json
import yaml
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import load_model
from sklearn.metrics import confusion_matrix

# =========================================
#  PATH CONFIGURATION
# =========================================
BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Genomic Disease Risk Predictor"
OUT_DIR = os.path.join(BASE_DIR, "outputs")
MODEL_PKL = os.path.join(OUT_DIR, "geneguard_model.pkl")
MODEL_H5 = os.path.join(OUT_DIR, "geneguard_model.h5")
META_YAML = os.path.join(OUT_DIR, "geneguard_metadata.yaml")
PRED_JSON = os.path.join(OUT_DIR, "prediction_results.json")

os.makedirs(OUT_DIR, exist_ok=True)

# =========================================
#  LOAD DATA + METADATA
# =========================================
with open(META_YAML, "r") as f:
    meta = yaml.safe_load(f)

pred_df = pd.read_json(PRED_JSON)
rf_model = joblib.load(MODEL_PKL)
dnn_model = load_model(MODEL_H5)

print("[INFO] Loaded models and prediction data.")
print("[INFO] Available columns:", pred_df.columns.tolist())

# =========================================
#  1 ACCURACY GRAPH (BAR CHART)
# =========================================
acc_rf = meta.get("model_randomforest_acc", 0)
acc_dnn = meta.get("model_dnn_acc", 0)

plt.figure(figsize=(6, 5))
plt.bar(["RandomForest", "DNN"], [acc_rf, acc_dnn], color=["skyblue", "lightcoral"])
plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy")
plt.ylim(0, 1)
for i, v in enumerate([acc_rf, acc_dnn]):
    plt.text(i, v + 0.02, f"{v:.2f}", ha='center', fontsize=12)
plt.tight_layout()
plt.savefig(os.path.join(OUT_DIR, "accuracy_graph.png"))
plt.show()

# =========================================
#  2 CONFUSION MATRIX HEATMAP (FIXED)
# =========================================
# Handle missing or mixed labels safely
pred_df = pred_df.dropna(subset=["TrueLabel", "PredLabel"], how="all")
true_labels = pred_df["TrueLabel"].fillna("Unknown").astype(str)
pred_labels = pred_df["PredLabel"].fillna("Unknown").astype(str)

unique_labels = sorted(set(true_labels.unique()) | set(pred_labels.unique()))

cm = confusion_matrix(true_labels, pred_labels, labels=unique_labels)

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap="YlGnBu",
            xticklabels=unique_labels, yticklabels=unique_labels)
plt.title("Confusion Matrix Heatmap")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.tight_layout()
plt.savefig(os.path.join(OUT_DIR, "confusion_heatmap.png"))
plt.show()

# =========================================
#  3 FEATURE CONCENTRATION GRAPH
# (Feature Importance Cumulative Distribution)
# =========================================
if hasattr(rf_model, "feature_importances_"):
    importances = np.sort(rf_model.feature_importances_)[::-1]
    cum_importance = np.cumsum(importances)
    plt.figure(figsize=(8, 5))
    plt.plot(range(1, len(importances) + 1), cum_importance, marker='o')
    plt.title("Feature Importance Concentration Curve")
    plt.xlabel("Number of Top Features")
    plt.ylabel("Cumulative Importance")
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.tight_layout()
    plt.savefig(os.path.join(OUT_DIR, "feature_concentration_graph.png"))
    plt.show()
else:
    print("[WARN] RandomForest feature_importances_ not found!")

# =========================================
#  4 PREDICTION GRAPH
# (Per-sample predicted vs true + probability intensity)
# =========================================
if "RF_Prob" in pred_df.columns:
    plt.figure(figsize=(10, 5))
    sns.scatterplot(
        data=pred_df,
        x=range(len(pred_df)),
        y="RF_Prob",
        hue=(pred_df["TrueLabel"] == pred_df["PredLabel"]),
        palette={True: "green", False: "red"},
        legend=False
    )
    plt.title("Prediction Probability Distribution per Sample")
    plt.xlabel("Sample Index")
    plt.ylabel("Predicted Probability (RF)")
    plt.tight_layout()
    plt.savefig(os.path.join(OUT_DIR, "prediction_graph.png"))
    plt.show()
else:
    print("[WARN] Column 'RF_Prob' not found in predictions JSON!")

# =========================================
#  SAVE SUMMARY JSON
# =========================================
summary = {
    "accuracy_graph": "accuracy_graph.png",
    "confusion_heatmap": "confusion_heatmap.png",
    "feature_concentration_graph": "feature_concentration_graph.png",
    "prediction_graph": "prediction_graph.png"
}

with open(os.path.join(OUT_DIR, "visualization_summary.json"), "w") as f:
    json.dump(summary, f, indent=4)

print("\n Visualization Complete! Saved graphs:")
for k, v in summary.items():
    print(f" - {v}")
```
